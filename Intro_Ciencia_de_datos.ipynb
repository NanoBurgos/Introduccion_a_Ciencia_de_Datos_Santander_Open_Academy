La **diferencia entre Bussiness Intelligence y Data Analytics** esta en el hechoi de que **Bussiness Intelligence** nos ayuda a tomar decisiones empresariales basadas en resultados pasados; mientras que **Data Analytics** nos ayuda a realizar predicciones que nos ayudaran en el futuro.

# **¿Que habilidades necesita un Cientifico de Datos?**

**Un Cientifico de Datos** es la persona responsable de limpiar, procesar y analizar los datos.

La limpieza de datos es de inmersa importancia en la ciencia de datos, ya que la calidad de los datos realmente determina la calidad del resultado del analisis

Tras la limpieza de los datos, los cientidficos de datos deben procesar los datos para poder obtenerlos en el formato correcto para un analisis mas detallado

Tanto los conocimientos tecnicos, como el saber transmitir sus hallazgos a traves del storytelling son muy importantes para este puesto

Un Cientifico de Datos debe de saber cuales son las metricas adecuadas para medir un fenomeno y como se calculan

# **Un Cientifico de Datos Competente**

Debe tener conocimientos de estadistica y diseño de algoritmo, asi como habilidades de Programacion y Comunicacion en combinacion con un profundo entendimiento empresarial

# **Un Analista de Datos** 

Debe poder identificar las areas problematicas de los datos y encontrar posibles soluciones

Deben limpiar y procesar los datos para que el algoritmo diseñado por los cientificos de datos pueda procesar estos datos correctamente

Un Analista de Datos debe poder visualizar los datos correctamente para que el resto de equipo los pueda entender

# **Al comparar las funciones de un Cientifico de Datos y un Analista de Datos**

El primero es responsable de diseñar e implementar el algoritmo correcto para lograr los resultados deseados

Mientras que el ultimo, es responsable de apoyar el esfuerzo de los cientificos de datos al ayudar a optimizar algunas tareas y procesos, como la limpieza y el procesamiento de datos

# **-----Estadística: la columna vertebral del análisis de datos-----**

# **Introducción al Big Data**

La estadistica es una rama de las Matematicas que se ocupa de la recopilacion, clasificacion, analisis e interpretacion de hechos numericos para extraer inferencias en funcion de su probabilidad cuantificable, es decir, su probabilidad de ocurrencia

# **La Estadistica se divide en tres disciplinas:**

**El Analisis Descriptivo:** tambien conocido como analisis exploratorio de datos, esto se refiere a la recopilacion, visualizacion y resumen de los datos

La probabilidad o la ciencia de la incertidumbre se refiere a las aplicaciones de alguna ley de la probabilidad para cuantificar la incertidumbre

La **Estadistica Inferencial:** o la ciencia de sacar conclusiones a partir de datos especificos, utilizando el conocimiento de la probabilidad
